```python
# Podemos armazenar o nome, idade, e dados clínicos de um paciente.

nome_paciente = "Ana"
idade = 45
pressao_sistolica = 120
pressao_diastolica = 80
glicose = 105

print("Nome:", nome_paciente)
print("Idade:", idade)
print("Pressão:", pressao_sistolica, "/", pressao_diastolica)
print("Glicose:", glicose)
```

## 2. Declaração Múltipla

```python
# Podemos declarar várias variáveis de uma só vez.
# Exemplo: peso e altura do paciente

peso, altura = 68, 1.70

print("Peso:", peso)
print("Altura:", altura)
```

## 4. Não se pode usar palavras reservadas como nome de variável

```python
# Python tem palavras especiais, chamadas “reservadas”,
# que não podem ser usadas como nome de variável.
# Exemplos: if, for, while, True, False, class, def...

# Exemplo errado:
# class = "Diabetes"  # Isso daria erro!

# Exemplo certo:
classe_diagnostico = "Diabetes Tipo 2"

print("Classe de diagnóstico:", classe_diagnostico)
```

## 5. Variáveis atribuídas a outras variáveis e ordem dos operadores

```python
# Podemos criar uma variável com base em outra.
# Exemplo: calcular a pressão média arterial (PAM)

pressao_media = (pressao_sistolica + 2 * pressao_diastolica) / 3

print("Pressão Sistólica:", pressao_sistolica)
print("Pressão Diastólica:", pressao_diastolica)
print("Pressão Média:", round(pressao_media, 1))
```

## 6. Operações com Variáveis

```python
# Vamos calcular o Índice de Massa Corporal (IMC)
peso = 68
altura = 1.70

imc = peso / (altura ** 2)

print("IMC:", round(imc, 2))

# Podemos também comparar resultados
if imc >= 25:
    print("Classificação: Sobrepeso")
else:
    print("Classificação: Normal")
```

## 7. Concatenação de Variáveis

```python
# Às vezes queremos combinar textos com números, como em relatórios médicos.

nome = "João"
idade = 60
diagnostico = "Hipertensão controlada"

# Jeito tradicional
print("O paciente " + nome + " tem " + str(idade) + " anos e diagnóstico de " + diagnostico + ".")

# Jeito moderno (f-string)
print(f"O paciente {nome} tem {idade} anos e diagnóstico de {diagnostico}.")
```
